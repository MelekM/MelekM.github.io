```python
import os
from PIL import Image

root_dir = '/Users/melekmizher/Desktop/MelekM.github.io/resources/photographs'

output_folder = '/Users/melekmizher/Desktop/MelekM.github.io/resources/thumbnails'

thumbnail_size = (200, 200)

# Function to create thumbnails if they don't already exist
def create_thumbnail(image_path, output_folder):
    base_name = os.path.basename(image_path)
    thumbnail_name = f'thumb_{base_name}'
    thumbnail_path = os.path.join(output_folder, thumbnail_name)

    # Check if the thumbnail already exists
    if os.path.exists(thumbnail_path):
        print(f"Thumbnail already exists for {base_name}, skipping.")
        return

    # Create the thumbnail if it doesn't exist
    try:
        img = Image.open(image_path)
        img.thumbnail(thumbnail_size)
        os.makedirs(output_folder, exist_ok=True)  # Create the output folder if it doesn't exist
        img.save(thumbnail_path)
        print(f"Thumbnail saved for {base_name}")
    except Exception as e:
        print(f"Error processing {image_path}: {e}")

# Loop through directories and create thumbnails
for foldername, subfolders, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.lower().endswith(('jpg', 'jpeg', 'png')):
            image_path = os.path.join(foldername, filename)
            create_thumbnail(image_path, output_folder)
```
